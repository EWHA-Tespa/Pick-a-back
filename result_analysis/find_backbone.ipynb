# Setting

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px         # pip install plotly 터미널에서 돌릴 것
import plotly.graph_objects as go
import plotly.io as pio
from plotly.subplots import make_subplots


pio.templates.default = 'plotly_dark' # 어두운모드 설정. 지우면 밝은 모드로 돌아옵니다.
```

```python
DATASETS = [
    # 'None',
    'aquatic_mammals',
    'fish',
    'flowers',
    'food_containers',
    'fruit_and_vegetables',
    'household_electrical_devices',
    'household_furniture',
    'insects',
    'large_carnivores',
    'large_man-made_outdoor_things',
    'large_natural_outdoor_scenes',
    'large_omnivores_and_herbivores',
    'medium_mammals',
    'non-insect_invertebrates',
    'people',
    'reptiles',
    'small_mammals',
    'trees',
    'vehicles_1',
    'vehicles_2' ]
```

## Data

```python
result_path = "../"
res = pd.read_csv(result_path + f'find_backbone_log.csv')
res
```

```python
result = {}
for i, task in enumerate(DATASETS):
    result[task] = res[res.task_id == (i+1)].reset_index()
```

# Result Analysis

## Corr. Matrix

```python
cos_matrix = res.pivot(index='task_id', columns='backbone', values='DDV cos-cos')

plt.figure(figsize=(10, 8))
sns.heatmap(cos_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('DDV cos-cos Heatmap')
plt.xlabel('Backbone Task')
plt.ylabel('Task ID')
plt.show()
```

```python
euc_matrix = res.pivot(index='task_id', columns='backbone', values='DDV euc-cos')

plt.figure(figsize=(10, 8))
sns.heatmap(euc_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('DDV euc-cos Heatmap')
plt.xlabel('Backbone Task')
plt.ylabel('Task ID')
plt.show()
```
